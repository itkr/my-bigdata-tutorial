```python
def tweets(*args, **kwargs):
    for tweet in mongo.twitter.sample.find(*args, **kwargs):
        if 'delete' not in tweet:
            yield {
                'created_at': tweet['created_at'],
                'text': tweet['text'],
            }
```

```python
pd.DataFrame(tweets({'lang': 'en', limit=2}))
```

```python
pd.DataFrame(tweets({'lang': 'en'}, limit=2))
```
